```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import Lasso
from sklearn.linear_model import Ridge
import math


data= pd.read_csv(r"C:\Users\HOLLARMYDE {ALEXIX}\Downloads\energydata_complete.csv")
```

```python
X = data['T2'].values.reshape(-1, 1)
y = data['T6'].values

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=1)
                                                    
l_model = LinearRegression()
l_model.fit(X_train, y_train)

y_pred = l_model.predict(X_test)
mse = mean_squared_error(y_test, y_pred)
rmse = math.sqrt(mse)

print(f"RMSE: {rmse:.3f}")
```

```python
data = data.drop(["date", "lights"], axis=1)
X = data.drop("Appliances", axis=1)
y = data["Appliances"]
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
l_model = LinearRegression()
l_model.fit(X_train_scaled, y_train)
y_train_pred = l_model.predict(X_train_scaled)
mae_train = mean_absolute_error(y_train, y_train_pred)

print(f"Mean Absolute Error (Training Set): {mae_train:.3f}")
```

```python
mse_train = mean_squared_error(y_train, y_train_pred)
rmse_train = (mse_train ** 0.5)

print(f"Root Mean Squared Error (Training Set): {rmse_train:.3f}")
```

```python
y_test_pred = l_model.predict(X_test_scaled)
mae_test = mean_absolute_error(y_test, y_test_pred)
print(f"Mean Absolute Error (Test Set): {mae_test:.3f}")
```

```python
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_test = (mse_test ** 0.5)
print(f"Root Mean Squared Error (Test Set): {rmse_test:.3f}")
```

```python
lasso_model = Lasso()
lasso_model.fit(X_train_scaled, y_train)
feature_weights = lasso_model.coef_
non_zero_feature_count = sum(feature_weights != 0)
print(f"Number of features with non-zero feature weights: {non_zero_feature_count}")
```

```python
lasso_y_test_pred = lasso_model.predict(X_test_scaled)
lasso_mse_test = mean_squared_error(y_test, lasso_y_test_pred)
lasso_rmse_test = (lasso_mse_test ** 0.5)
print(f"RMSE with Lasso Regression (Test Set): {lasso_rmse_test:.3f}")

```

```python
y_train_pred = l_model.predict(X_train_scaled)
y_test_pred = l_model.predict(X_test_scaled)
mse_train = mean_squared_error(y_train, y_train_pred)
mse_test = mean_squared_error(y_test, y_test_pred)
rmse_train = (mse_train ** 0.5)
rmse_test = (mse_test ** 0.5)
print(f"RMSE on Training Set: {rmse_train:.3f}")
print(f"RMSE on Test Set: {rmse_test:.3f}")
if rmse_train < rmse_test:
    print("The model may be overfitting to the training set.")
else:
    print("The model is not showing significant signs of overfitting.")
```

```python
r_model = Ridge()
r_model.fit(X_train_scaled, y_train)
from sklearn.metrics import mean_squared_error
ridge_y_test_pred = r_model.predict(X_test_scaled)
ridge_mse_test = mean_squared_error(y_test, ridge_y_test_pred)
ridge_rmse_test = (ridge_mse_test ** 0.5)
print(f"RMSE with Ridge Regression on Test Set: {ridge_rmse_test:.3f}")
original_rmse_test = rmse_test
print(f"RMSE with Original Linear Regression on Test Set: {original_rmse_test:.3f}")
if ridge_rmse_test == original_rmse_test:
    print("The RMSE values are the same for both models.")
elif ridge_rmse_test < original_rmse_test:
    print("The RMSE is lower with the Ridge regression model, indicating potential improvement.")
else:
    print("The RMSE is higher with the Ridge regression model.")


```
